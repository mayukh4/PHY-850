Question 1 (/5)
===

Using library scipy.stats, plot the pdf and cdf of a gaussian distribution with $\mu=5$ and $\sigma=3$.  Generate $10^4$ samples, and plot the distribution in counts per bin, overlaying the expected deistribution determined from the pdf.

```python
# importing the libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# writing down the given data
mean =5
sd = 3

# generating the data for the pdf 
xdata1= np.linspace(-10,20,100)

# generating the data for the cdf 
xdata2= np.linspace(-10,20,100)

# generating 10^4 data samples
data = np.random.randn(10000)

# generating the data for the PDF distribution
xdata3= np.linspace(-4,4,100)


#making all the plots
fig, axs = plt.subplots(1,3,figsize=(21,4))

axs[0].set_title("PDF of a gaussian distribution with $\mu=5$ and $\sigma=3$",size=12)
axs[0].plot(xdata1, norm.pdf(xdata1, mean, sd))
axs[0].set_ylabel("f(x)",size=15)
axs[0].set_xlabel("x",size=15)


axs[1].set_title("CDF of a gaussian distribution with $\mu=5$ and $\sigma=3$",size=12)
axs[1].plot(xdata2, norm.cdf(xdata2, mean, sd))
axs[1].set_xlabel("x",size=15)
axs[1].set_ylabel("f(x)",size=15) 

axs[2].plot(xdata3,norm.pdf(xdata3),'m-',label="PDF")
axs[2].hist(data,bins=np.linspace(-5,5,100),density=True,label='counts per bin')
axs[2].set_title("PDF of random distribution : Gaussian",size=12)
axs[2].set_ylabel("Probability density",size=12)
axs[2].set_xlabel("x",size=12)
axs[2].legend()


plt.show()
```

Question 2 (/5)
===
Simulate the sum of two Poisson variables with respective means $3$ and $5$ and show that it behaves as expected in terms of mean and variance.

```python
# importing the libraries

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson

x = poisson.rvs(3, size=1000) # first poisson variable with mean 3
y = poisson.rvs(5, size=1000) # 2nd possion variable with mean 5

z = x + y # sum of two poisson variavle : is also a poisson variable

fig, axs = plt.subplots(1,3,sharex=True,sharey=True,figsize=(18,5))


axs[0].set_title("Distribution of Poisson Variable X($ \lambda = 3 $)")
axs[0].hist(x,density=True)
axs[0].set_ylabel("Probability Density",size=12)
axs[0].set_xlabel("Poisson Variable X",size=12)


axs[1].set_title("Distribution of Poisson Variable Y($ \lambda = 5 $)")
axs[1].hist(y,density=True)
axs[1].set_xlabel("Poisson Variable Y",size=12)
#axs[1].set_xlabel("$ X^-1 $",size=12)


axs[2].set_title("Distribution of sum of Poisson Variable Z($ Z= X+Y $) and ($ \lambda = 8 $)")
axs[2].hist(z,density=True)
axs[2].set_xlabel("Poisson Variable Z",size=12)
#axs[2].set_xlabel("$ X^-1 $",size=12)

plt.show()

# -------------------------NOTE-----------------------------
# As expected the mean and variance of the third possion variable which is the sum of the two is sum of mean and variance of each of the independent poisson variable.
```

Question 3 (/5)
===
Analytically determine the distribution of the inverse of a normal variable, and compare to simulations.

```python
# importing the libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import norminvgauss

# generating the random data

data = norm.rvs(size=10000)

# plot the dist of the inverse of the variables

invdata= [1.0/d for d in data]

# using a built in scipy library to simulate the same situation

a, b = 1, 0.8  # a is the tail-heaviness of the inverse normal distribution and b is the assymetry parameter
r = norminvgauss.rvs(a, b, size=10000)

#plotting the data

fig, axs = plt.subplots(1,3,sharex=True,sharey=True,figsize=(19,4))

axs[0].hist(data,bins=np.linspace(-5,5,100),density=True)
axs[0].set_title("Distribution of a normal variables")
axs[0].set_xlabel("X",size=12)

axs[1].hist(invdata,bins=np.linspace(-5,5,100),density=True)
axs[1].set_title("Distribution of inverse of a normal variables")
axs[1].set_xlabel("$ X^-1 $",size=12)

axs[2].hist(r,bins=np.linspace(-5,5,100),density=True)
axs[2].set_title("Simulating the inverse of normal variables with inbuilt function")
axs[2].set_xlabel("$ X^-1 $",size=12)

fig.text(0.04, 0.5, 'Probability Density', va='center', rotation='vertical',size=12)
plt.show()

# -------------------------NOTE-----------------------------
# the most frequently occuring bin size is the mode. In this distribution the mode is 0 (which is the mean too), hence when we take the inverse (i.e inverse of 0) we get infinitely large terms!
```

Question 4 (/5)
===
Generate $10^4$ random variables $X$, each of which is the sum of $m$ uniform variables over $0$--$1$.  Repeat for $m$ in 1, 2, 3, 10, 100, 1000.   At each stage, plot the distribution, and determine the mean and standard deviation of $X$.  What does this illustrate?

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt

# # generating 10^4 data samples
# data = np.random.randn(10000)

# number of sample
num = [1,2,3,10,100,1000] 

# list of sample means
means = []

# Generating 1,2,3,10,100,1000 random numbers from 0 to 1
# taking their mean and appending it to list means.
for j in num:
    
    x = [np.mean(
        np.random.randint(
            -50, 50, j)) for _i in range(10000)] # generating 10^4 random samples each of which is a sum of m uniform variables 
    means.append(x)
k = 0

# plotting all the means in one figure
fig, ax = plt.subplots(2, 3, figsize =(13, 8))
for i in range(0, 2):
    for j in range(0, 3):
        # Histogram for each x stored in means
        ax[i, j].hist(means[k],bins=50, density = True)
        ax[i, j].set_title(label = 'm='+str(num[k]))
        k = k + 1
plt.show()

# -------------------------NOTE-----------------------------
#clearly with increasing the number of the uniform variable m we can see how it starts to behave closely to a Gaussian
```
